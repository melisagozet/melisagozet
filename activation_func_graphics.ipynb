```python
def derivative(x): # Derivative of ReLU
  d = np.array(x, copy=True)
  d[x < 0] = 0
  d[x >= 0] = 1
  return d
```

```python
def ReLU(x):
  y = np.maximum(0, x)
  return y
```

```python
x = np.linspace(-6, 6, 50)

plt.plot(x, ReLU(x), linewidth=2.8, label="ReLU")
plt.plot(x, derivative(x), linewidth=2.5, label="Derivative")

plt.legend()
plt.xlabel("x")
plt.ylabel("ReLU(X)")
  
plt.show()
```

```python
def tanh(x): # tanh function
  return 2*sigmoid(2*x) - 1
def derivative(x): # derivative of tanh
  return 1-tanh(x)**2
```

```python
x = np.linspace(-6, 6, 50)
z = 1/(1 + np.exp(-x))

plt.plot(x, tanh(x), linewidth=2.8, label="Tanh(x)")
plt.plot(x, derivative(x), linewidth=2.5, label="Derivative")

plt.legend()
plt.xlabel("x")
plt.ylabel("Tanh(x)")
  
plt.show()
```

```python
from matplotlib import pyplot

#Exponential Linear Squashing Activation
def elish(x):

  if x>=0:
    return x/(1+np.exp(-x))
  
  else:
    return (np.exp(x)-1)/(1+np.exp(-x))

# define a series of inputs
series_in = [x for x in range(-10, 11)]
# calculate outputs for our inputs
series_out = [elish(x) for x in series_in]
# line plot of raw inputs to rectified outputs
pyplot.plot(series_in, series_out)
pyplot.show()
```
